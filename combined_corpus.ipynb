```python
# Load the Excel file
filtered_corpus_path = 'C:\\Users\\eugen\\corpus_article_debt\\filtered_corpus.xlsx'
df = pd.read_excel(filtered_corpus_path)

print(f"Loading titles from: {filtered_corpus_path}")
titles = set(df['Title']) 
print(f"Total titles loaded: {len(titles)}")

# Source and target directories
source_directories = {
    'German': 'C:\\Users\\eugen\\corpus_article_debt\\raw files\\raw docs DE',
    'French': 'C:\\Users\\eugen\\corpus_article_debt\\raw files\\raw docs FR'
}
target_directory = 'C:\\Users\\eugen\\corpus_article_debt\\raw files\\processed_docs'

# Ensure the target directory exists
os.makedirs(target_directory, exist_ok=True)
print(f"Target directory is set to: {target_directory}")
```

```python
# Define the similarity function for comparison
def similar(a, b):
    """Return the ratio of similarity between two strings."""
    return SequenceMatcher(None, a, b).ratio()

total_matched_files_count = 0
```

```python
# Process each source directory
for language, source_directory in source_directories.items():
    matched_files_count = 0
    print(f"Processing {language} documents in {source_directory}")

    for filename in os.listdir(source_directory):
        file_title_without_ext = os.path.splitext(filename)[0]

        # Direct comparison without normalization
        if any(similar(file_title_without_ext, title) >= 0.8 for title in titles):
            source_path = os.path.join(source_directory, filename)
            target_path = os.path.join(target_directory, filename)
            shutil.copy2(source_path, target_path)
            print(f"Matched and copied: {filename}")
            matched_files_count += 1
        else:
            print(f"No match for: {filename}")

    print(f"Total matched and copied {language} files: {matched_files_count}")
    total_matched_files_count += matched_files_count

print(f"Overall total matched and copied files: {total_matched_files_count}")
```
